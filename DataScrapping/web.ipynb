```python
import requests
import time
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import json
import os
```

URL -> HTML EXTRACTION

```python
basicUrl = "https://www.shiksha.com/university/galgotias-university-greater-noida-37105"
FOLDER_NAME = 'GALGOTIAS-Delhi'
mapping = {
    0: "college-info",
    1: "courses",
    2: "fees",
    3: "reviews",
    4: "admission",
    5: "placement",
    6: "cutoff",
    7: "ranking",
    8: "infrastructure",
    9: "faculty"
}

if not os.path.exists(FOLDER_NAME):
    os.makedirs(FOLDER_NAME)
    print("Directory ", FOLDER_NAME, " Created ")
```

```python
url = []
'''
    [College Info , Courses , Fees , Reviews , Admissions , Placements , Cutoffs , Rankings , Infrastructure , Faculty]
'''
url.append(basicUrl)
url.append(basicUrl + "/courses")
url.append(basicUrl + "/fees")
url.append(basicUrl + "/reviews")
url.append(basicUrl + "/admission")
url.append(basicUrl + "/placement")
url.append(basicUrl + "/cutoff")
url.append(basicUrl + "/ranking")
url.append(basicUrl + "/infrastructure")
url.append(basicUrl + "/faculty")

print("URL : " + url[0])
```

```python
'''
    extracting html from the url and saving it
'''

for i in range(len(url)):
    
    print(url[i])
    headers = {
        'referer': url[i],
        'sec-ch-ua': '"Not A(Brand";v="8", "Chromium";v="132", "Google Chrome";v="132"',
        'sec-ch-ua-mobile': '?0',
        'sec-ch-ua-model': '""',
        'sec-ch-ua-platform': '"Windows"',
        'sec-ch-ua-platform-version': '"19.0.0"',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36'
    }

    ua = UserAgent()
    headers['user-agent'] = ua.random

    time.sleep(3)
    session = requests.Session()
    r = session.get(url[i], headers=headers)

    DIR_NAME = FOLDER_NAME + '/' + mapping[i] + ".html"
    with open(DIR_NAME, 'w', encoding='utf-8') as f:
        f.write(r.text)
    print(f"Saved {mapping[i]} page")
    f.close()
    print("Done for ", mapping[i])
```

HTML TO JSON CONVERSION



```python
FOLDER_NAME = "AMITY-GURUGRAM-Delhi"

PAGE_INFO = []

'''
    [College Info , Courses , Fees , Reviews , Admissions , Placements , Cutoffs , Rankings , Infrastructure , Faculty]
'''
```

```python
colleges = next(os.walk('.'))[1]
colleges.remove("JSON")
print(colleges)
```

```python

for j in colleges:
    for i in os.listdir(j):
        print(i)
        with open(FOLDER_NAME + '/' + i, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')
            f.close()
        PAGE_INFO.append(soup.find("body"))
```

```python
li = [PAGE_INFO[0].find_all("div")]
print(PAGE_INFO[0].prettify())

```

```python
'''
faq__according-wrapper
'''

page_info = PAGE_INFO[0].find_all("div", class_="faq__according-wrapper")
```

```python
table = page_info[1].find_all("table" , {"border" : "1"})
print(table[0].prettify())
```

```python
'''
    Amity Noida Admission 2025: Eligibility & Selection Criteria
Amity University Noida offers various UG & PG courses in various specialisations. To get admission to Amity Noida, candidates must pass Class 12 for UG courses and graduation for PG courses. Check below the eligibility and selection criteria for top courses related to Amity University Noida course admissions:
'''
```

```python
'''
    get element by className "faq__according-wrapper"
'''

faq_according_wrapper = soup.find_all('div', class_='faq__according-wrapper')
```

```python
'''
    Table is left for now 
    major info for json object:
    introduction : [intro_1 , intro_2] ;
    nirf_rank ; 
    nirf_category ;
    admission_details ;
'''


nirfTag = (faq_according_wrapper[0].find('div')).find('p').find_all('strong')
nirf_rank = nirfTag[0].text
nirf_category = nirfTag[1].text
nirf_rank = nirf_rank.strip('#')
nirf_category = nirf_category.strip("'")


intro_1 = (faq_according_wrapper[1].find('div')).find('p').text
intro_2 = (faq_according_wrapper[1].find('div')).find_all('p')[1].text
intro_2 = intro_2.split(".")[:-1]
intro_2 = ". ".join(intro_2)

adminssion_details = faq_according_wrapper[2].find('div').find('p').text
adminssion_details = adminssion_details.split('.')[:-1]
adminssion_details = '.'.join(adminssion_details)

course_offer = faq_according_wrapper[3].find('div').find('p').text
course_offer = course_offer.split('.')[:-1]
course_offer = '.'.join(course_offer)
```

```python
print(faq_according_wrapper[3].find('div'))
```

```python
data = {
    "introduction": [intro_1, intro_2],
    "nirf_rank": nirf_rank,
    "nirf_category": nirf_category,
    "admission_details": adminssion_details,
    "course_offer": course_offer
}


json_data = json.dumps(data, indent=4)

with open("amity-university-delhi_shiksha.json", "w") as file:
    file.write(json_data)

print("JSON Data Saved Successfully!")
```
